```python
from pyspark import SparkConf, SparkContext 

conf = SparkConf() 
sc = SparkContext(conf=conf)
```

```python
x = "MWűGkyösdwinko Qau fSJpDaTrykv!w"

rdd = sc.parallelize([x[i:i+2] for i in range(0, len(x), 2)])
res = rdd.map(lambda x: x[0]).collect()

print("".join(res))
```

```python
szamok = sc.parallelize(range(1000))
```

```python
szamok.collect()
```

```python
szamok.take(10)
```

```python
szamok.count()
```

```python
def paros(szam):
    if szam % 2 == 0:
        return True
    else:
        return False

szamok.filter(paros).take(10)
```

```python
szamok.filter(lambda x: x % 2 == 0).take(10)
```

```python
def osszeg(a,b):
    c = a + b
    return c

szamok.filter(lambda x: x % 2 == 0).reduce(osszeg)
```

```python
text = sc.textFile('alkotmany.txt')
```

```python
text.take(2)
```

```python
keresendo = "magyar"

text.filter(lambda line: keresendo in line).collect()
```

```python
text.map(lambda line: line.split(" ")).take(2)
```

```python
text.flatMap(lambda line: line.split(" ")).sortBy(lambda x: x).take(2)
```

```python
text.flatMap(lambda line: line.split(" ")).map(lambda word: (word.lower(), 1)).reduceByKey(lambda a, b: a+b).sortBy(lambda t: t[1], False).take(10)
```
